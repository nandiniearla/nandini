```python
data = {
    'RawData': [
        '12345678     12301 45                                                  9',
        '00000001     45601 78                                                  0',
        '1234abcd     xyz99 01                                                  X',
        '87654321     78901 12                                                  1',
    ]
}
df = pd.DataFrame(data)
# print(df)
def extract_fields(row):
    record = row['RawData']
    return pd.Series({
        'section number': record[0:8],
        'Blank': record[8:13],
        'Branch': record[13:16],
        'Index Number': record[16:18],
        'Market Division': record[18:20],
        'Filler': record[20:79],
        'Transaction Code': record[79]
    })
df_extracted = df.apply(extract_fields,axis=1) 

def validate_fields(row):
    errors = []
    if not row['section number'].strip().isdigit():
        errors.append('Invalid section number')
    if row['Blank'].strip() != '':
        errors.append('Blank not blank')
    if not row['Branch'].strip().isdigit():
        errors.append('Invalid Branch')
    if row['Index Number'] != '01':
        errors.append('Invalid Index Number')
    if not row['Market Division'].strip().isdigit():
        errors.append('Invalid Market Division')
    if row['Filler'].strip() != '':
        errors.append('Filler not blank')
    if not row['Transaction Code'].isdigit():
        errors.append('Invalid Transaction Code')
    return ', '.join(errors) if errors else 'Valid'
df_extracted['validation result'] = df_extracted.apply(validate_fields,axis=1)
print(df_extracted)
```
