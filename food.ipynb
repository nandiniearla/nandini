```python
def prime(n):
    if n<=1:
        return 0
    m=1
    while m<=n:
        k=m
        i=1
        c=0
        while i<=k:
            if k%i == 0:
                c+=1
            i+=1
        m+=1
        if c == 2:
           print(k,end =' ')
prime(100)
```
