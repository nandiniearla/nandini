```python
class Profile:
    def _init_(self, name, age, gender, occupation, interests):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation
        self.interests = interests

    def display_profile(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Occupation: {self.occupation}")
        print(f"Interests: {', '.join(self.interests)}")
        print("-" * 50)

    @classmethod
    def from_input(cls):
        name = input("Enter your name: ")
        age = int(input("Enter your age: "))
        gender = input("Enter your gender: ")
        occupation = input("Enter your occupation: ")
        interests = []
        while True:
            interest = input("Enter an interest (blank to stop): ")
            if interest == "":
                break
            interests.append(interest)
        return cls(name, age, gender, occupation, interests)


class MatrimonialSite:
    def _init_(self):
        self.profiles = []

    def create_profile(self):
        profile = Profile.from_input()
        self.profiles.append(profile)
        print("Profile created successfully!")

    def browse_profiles(self):
        if len(self.profiles) > 0:
            print("Available Profiles:")
            for profile in self.profiles:
                profile.display_profile()
        else:
            print("No profiles available.")

    def search_profiles(self, search_criteria):
        found_profiles = []
        for profile in self.profiles:
            if search_criteria.lower() in profile.interests:
                found_profiles.append(profile)
        if len(found_profiles) > 0:
            print(f"Found {len(found_profiles)} profile(s) matching the search criteria:")
            for profile in found_profiles:
                profile.display_profile()
        else:
            print("No profiles found matching the search criteria.")


# Example usage
site = MatrimonialSite()
 
while True:
    print("Matrimonial Site")
    print("1. Create Profile")
    print("2. Browse Profiles")
    print("3. Search Profiles")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        site.create_profile()
    elif choice == "2":
        site.browse_profiles()
    elif choice == "3":
        search_criteria = input("Enter search criteria: ")
        site.search_profiles(search_criteria)
    elif choice == "4":
        break
    else:
        print("Invalid choice. Please try again.")
```
