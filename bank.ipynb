```python
import datetime
class Bank:
    bname="INDIAN Bank"
    branch="ameerpet"
    branch_code=320
    ifsc='IND000N420'
    toll_no:858652209
    roi=7
    no_cust=0
    cust_details={}
    cust_transcations={}
    def __init__(self,name,age,email,phone,address,adhaar,pancard,dob,gender,pin,bal=0):
        self.cname=name
        self.age=age
        self.email=email
        self.phone=phone
        self.address=address
        self.adhaar=adhaar
        self.pancard=pancard
        self.dob=dob
        self.gender=gender
        self.dob=dob
        self.pin=pin
        self.bal=bal
        self.no_cust_inc()
        self.account_no=100+self.no_cust    
        self.concat_cust_details(self.account_no,self)

    @classmethod
    def no_cust_inc(cls):
        cls.no_cust+=1

    @classmethod
    def concat_cust_details(cls,acc_no,obj):
        cls.cust_details[acc_no]=obj
    @classmethod
    def change_ROI(cls):
        cls.roi=int(input("enter the ROI:"))
        print("ROI is changed")

    def check_Bal(cls):
        acc_no=int(input("enter the account no: "))
        pin=int(input("enter the account pin no: "))
        for i in cls.cust_details:
            res=cls.cust_details[i]
            if res.account_no==acc_no and res.pin==pin:
                print(res.account_no,"balance is:",res.bal)
                return
        else:
            print("invalid account_no or pin")

    @classmethod
    def concat_cust_tran(cls,acc_no,data):
        if acc_no in cls.cust_transcations:
            cls.cust_transcations[acc_no]+=[data]
            print(acc_no,data)
        else:
            cls.cust_transcations[acc_no]=[data]
            print(acc_no,data)
    
    def deposit(self,count=0):
        if count==3:
            return
        amt=int(input("enter the deposit amount:"))
        if amt>=50:
            self.bal+=amt
            self.sucess()
            self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(),'type':'deposit','amount':amt,'bal':self.bal})
            return
        else:
            print("amount should be mimimum 50 rps")
            self.failed()
            return self.deposit(count+1)
        
    def withdraw(self,count=0):
        if count==3:
            return
        amt=int(input("enter the withdraw amount:"))
        if amt>=100 and self.bal>=amt:
            self.bal-=amt
            self.sucess()
            self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(),'type':'withdraw','amount':amt,'bal':self.bal})
            self.sucess()
            return
        else:
            print("amount should be minimum 100 rps")
            self.failed()
            return self.withdraw(count+1)
    def mini_statement(self):
        print('Date'.center(25),'Type'.center(15),'Aomunt'.center(15),'Balance'.center(15),sep='|')
        print(80*'-')
        if self.account_no in self.cust_transcations:
            for i in self.cust_transcations[self.account_no]:
                print(str(i['date']).center(25),str(i['type']).center(15),str(i['amount']).center(15),str(i['bal']).center(15),sep=" | ")
                print(80*'-')
        else:
            print("no money")
    
    def check_bal(self):
        print("your current balance:",self.bal)
    def pin_change(self):
        old=int(input("enter the old pin") )
        if self.pin==old:
            self.pin=int(input("enter the new pin:"))
            print("pin changed")
            self.sucess()
        else:
            self.failed()
    
    def cust_update(self):
        n=int(input("choose the option:\n1.name\n2.phone\n3.email\n4.address\n5.exit"))
        match n:
            case 1:
                self.cname=input("enter the name")
            case 2:
                self.phone=int(input("enter the phone no"))
            case 3:
                self.email=input("enter the email")
            case 4:
                self.address=input("enter the address")
            case _:
                return
        return self.cust_update()
    
    def account_details(self):
        print("Account no:", self.account_no)
        print("name:", self.cname)
        print("Age:",self.age)
        print("Email:",self.email)
        print("phone:",self.phone)
        print("Balance:",self.bal)
        print("IFSC:",self.ifsc)
        print("pan:",self.pancard)
        print("Adhaar:",self.adhaar)
        print("Address:",self.address)

    def transfer(self):
        print("your current balance:",self.bal)
        acc_no=int(input("enter the Benificiary Accno:"))  
        ifsc_code=input("enter the Benificiary IFSC:")
        if acc_no in self.cust_details and ifsc_code== self.cust_details[acc_no].ifsc:
            amt=int(input("enter the Transfer amount:"))
            if self.bal>=amt:
                other=self.cust_details[acc_no]
                other.bal+=amt
                self.bal-=amt
                self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(),'type':'Transfer(w)','amount':amt,'bal':self.bal})
                self.concat_cust_tran(self.account_no,{'date':datetime.datetime.now(),'type':'Transfer(d)','amount':amt,'bal':self.bal})

                self.sucess()
                print("After transfer current balance",self.bal)
                return
            else:
                self.failed()
                print("insufficient balance")
        else:
            print("Invalid Account number or Ifsc code")
        return   
    @classmethod
    def home(cls,self):
        n=int(input("choose the option:\n0.AccountDetails\n1.check bal\n2.Deposit\n3.withdraw\n4.mini statement\n5.update\n6.pin change\n7.check ROI\n8.Transfer\n9.Exit"))
        print('*'*150)
        match n:
            case 0:
                self.account_details()
            case 1:
                self.check_bal()
            case 2:
                self.deposit()
            case 3:
                self.withdraw()
            case 4:
                self.mini_statement()
            case 5:
                self.cust_update()
            case 6:
                self.pin_change()
            case 7:
                print("Today Rate of Interset",self.roi)
            case 8:
                self.transfer()
            case _:
                return 
        return cls.home(self)
    @staticmethod
    def sucess():
        print("|---------------------|")
        print('|Transcation is sucess|')
        print("|---------------------|")

    @staticmethod
    def failed():
        print("|---------------------|")
        print("|Transcation is Failed|")
        print("|---------------------|")
    
obj=Bank('nandini',22,'nandini@gmail.com',95637807342,'hyd',245218556656,'abcdg67dfb','2001-06-08','Female',2222,1000)
obj1=Bank("srihari",20,'srihari@gmail.com',9623569975,'hyd',234512678972,'webghg3456b','2001-06-10','Female',1111)
def main(cls):
    acc_no=int(input("enter the account_no:"))
    pin=int(input("enter the pin_no"))
    if acc_no in cls.cust_details and cls.cust_details[acc_no].pin==pin:
        self=cls.cust_details[acc_no]
        cls.home(self)
    else:
        print("invalid account no or pin")
main(Bank)
```
